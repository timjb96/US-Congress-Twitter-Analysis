```python
import pandas as pd
import pathlib as path
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_path = "./Resource/housesessions109_118.csv"
congress_df = pd.read_csv(file_path)
congress_df.head()
```

```python
# Create dataframe for the 109 Congress specifically
congress109_df = congress_df[congress_df['CongressSession'] == 109]
congress109_df
```

```python
congress109_df.dtypes
```

```python
# Party Color references: 
# Republican: E81B23
# Democratic: 00AEF3

# Create Histogram by Gender and Policitcal Party
congress109_df.hist(column='AgeWhenSessionStarted', 
                    by=['gender', 'party'],
                    bins=10, 
                    color='#00AEF3', 
                    figsize=(10,7))

# fig.save
plt.savefig('congress109.png')
```
